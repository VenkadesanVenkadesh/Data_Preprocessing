Missing value percentage

```python
col = df1.columns

def missing_percentage(a):
  return df1[a].isnull().sum() / df1[a].size*100
for i in col:
  val = missing_percentage(i)
  if val >= 30:               # null value percentage more than 30% column will deleted
    del df1[i]
  elif val == 0:
    pass
  else:
    print(i,'---',val)


```

```python
df1.drop('Description',axis = 1, inplace = True)
```

```python
df1.drop('Permit Number',axis = 1, inplace = True)
```

```python
df1.drop(labels = ['Issued Date','First Construction Document Date','Permit Expiration Date'],axis = 1, inplace = True)
```

Fillna for numerical data

```python
df1['Proposed Construction Type'] = df1['Proposed Construction Type'].fillna(df1['Proposed Construction Type'].mean())
```

```python
num_col = ['Number of Proposed Stories','Number of Existing Stories','Estimated Cost','Revised Cost','Existing Units','Proposed Units','Plansets','Existing Construction Type','Supervisor District']
for i in num_col:
  df1[i] = df1[i].fillna(df1[i].mean())
```

Fillna for catogrical value


```python
df1['Street Suffix'] = df1['Street Suffix'].fillna(df1['Street Suffix'].mode()[0])
```

```python
catg_col = ['Existing Use','Proposed Use','Existing Construction Type Description','Neighborhoods - Analysis Boundaries','Proposed Construction Type Description']
for i in catg_col:
  df1[i] = df1[i].fillna(df1[i].mode()[0])
```

```python
df1.dropna(inplace = True) # for athor column like location
```

```python
objt = ['Block','Lot','Street Name','Street Suffix','Current Status','Current Status Date','Filed Date','Existing Use','Proposed Use','Existing Construction Type Description','Proposed Construction Type Description','Neighborhoods - Analysis Boundaries','Location']
for i in objt:
  df1[i] = le.fit_transform(df1[i])
```
